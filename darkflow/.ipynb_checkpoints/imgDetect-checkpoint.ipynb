```python
import cv2
from darkflow.net.build import TFNet
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'svg'
```

```python
options = {
    'model':'cfg/yolo.cfg',
    'load':'bin/yolov2.weights',
    'threshold':0.3
    
}


tfnet = TFNet(options)
```

```python
img = cv2.imread('img.jpeg',cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

result = tfnet.return_predict(img)
result
```

```python
top_left = (result[0]['topleft']['x'], result[0]['topleft']['y'])
btm_right = (result[0]['bottomright']['x'], result[0]['bottomright']['y'])
label = result[0]['label']

img = cv2.rectangle(img, top_left, btm_right, (0,0,255), 4)

font = cv2.FONT_HERSHEY_SIMPLEX
img = cv2.putText(img, label, top_left,font, 0.7, (0,0,0), 2)
plt.imshow(img)
plt.show()
```
