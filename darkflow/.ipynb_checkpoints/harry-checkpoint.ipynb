```python
import cv2
from darkflow.net.build import TFNet
import matplotlib.pyplot as plt
```

```python
options = {
    'model':'cfg/yolo.cfg',
    'load':'bin/yolov2.weights',
    'threshold':0.3
    
}


tfnet = TFNet(options)
```

```python
img = cv2.imread('harry.jpg',cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

results = tfnet.return_predict(img)
results
```

```python
for result in results:
    top_left = (result['topleft']['x'], result['topleft']['y'])
    btm_right = (result['bottomright']['x'], result['bottomright']['y'])
    label = result['label']
    img = cv2.rectangle(img, top_left, btm_right, (0,0,255), 7)

    font = cv2.FONT_HERSHEY_COMPLEX
    img = cv2.putText(img, label, top_left,font, 2, (180,210,80), 10)
plt.figure(figsize=(20,10))    
plt.imshow(img)
plt.show()
```
